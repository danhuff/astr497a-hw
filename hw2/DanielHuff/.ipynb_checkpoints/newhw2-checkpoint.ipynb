```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from astropy.io import fits
```

```python
dat = fits.getdata('hw2_danhuff.fit')
```

```python
blue=dat[dat['r'] <= 20.66]
green=dat[(dat['r'] > 20.66) & (dat['r'] <= 21.33)]
red=dat[dat['r'] > 21.33]
```

```python
decbins=2*(dat['dec'].max()-dat['dec'].min())
rabins=2*(dat['ra'].max()-dat['ra'].min())

Hb_B, xedges_B, yedges_B = np.histogram2d(blue['dec'],blue['ra'],[decbins,rabins])
Hb_G, xedges_G, yedges_G = np.histogram2d(green['dec'],green['ra'],[decbins,rabins])
Hb_R, xedges_R, yedges_R = np.histogram2d(red['dec'],red['ra'],[decbins,rabins])

#from paper - trim dim & bright bins
Hb_R[np.where((Hb_R < 25.5) | (Hb_R > 82.5))]=0
Hb_G[np.where((Hb_G < 26.75) | (Hb_G > 76))]=0
Hb_B[np.where((Hb_B < 24.5) | (Hb_B > 66.75))]=0
```

```python
d=np.dstack([Hb_R/Hb_R.max(),Hb_G/Hb_G.max(),Hb_B/Hb_B.max()])

f,ax=plt.subplots(figsize=[18,8])
ax.imshow(np.flipud(np.fliplr(d)),extent=(yedges_R[-1],yedges_R[0],xedges_R[0],xedges_R[-1]),aspect='equal',origin='upper')
ax.set_xlabel(r'$\alpha$ (deg)',size=16)
ax.set_ylabel(r'$\delta$ (deg)',size=16)
ax.set_title('SDSS False Color Plot (20 < r < 22) & (g-r < 0.4)',size=16)
```
